```python
import requests
from bs4 import BeautifulSoup

url = "https://www.seoul.go.kr/main/index.jsp"

response = requests.get(url)

# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.text, "html.parser")
print(soup.title)
print(soup.title.text)
```

```python
# 메뉴 제목 추출 - select_one() : CSS 선택자 이용하여 요소 하나 추출
first_li = soup.select_one("div.m_service > ul > li")
print(first_li.text)
```

```python
# 모든 메뉴 제목 추출 - select() : CSS 선택자 이용하여 요소 여러 개 추출
all_li = soup.select("div.m_service > ul > li")
print(all_li) # 요소 객체 리스트

for li in all_li:
    print(li.text)
```
